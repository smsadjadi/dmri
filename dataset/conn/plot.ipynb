```python
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
data_dir = "/home/ubuntu/Github/nitest/dmri/dataset/subj_01/analyzed_fsl/probtrackx/fdt_network_matrix"
Conn = np.loadtxt(data_dir, dtype=float)  
```

```python
fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow( Conn, cmap="viridis", origin="upper", aspect="equal")
cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Weight / strength")
ax.set_title("Connectivity matrix")
ax.set_xlabel("Target node / column")
ax.set_ylabel("Source node / row")
plt.tight_layout()
plt.show()
```
